# Preprocessing EBSSA data

I named it gen4 but is is actually closer to gen1 data...

Commands for preprocessing EBSSA data:

python convert_mat_to_gen4.py (EBSSA -> ebssa_gen4)

python copy_paired_gen4.py (ebssa_gen4 -> gen4_full_video)

python count_objects_gen4.pt (misc)

python split_gen4_into_60s.py (gen4_full_video -> gen4_60s)

python split_gen4_60s_train_eval_test.py (gen4_60s -> gen4)

python3 scripts/data/psee_to_frames.py data/gen4 data/gen4_psee -n 10 --mangling rvt-gen1 -d 50ms --workers 1 (preprocessing from prophesee)

* While preprocessing the data, a simple change has been done in psee_to_frames.py found in scripts/data.
* Added a fixed size of considering the bigger sensor size to make the npz frames all the same size.


Copy and past:

```python
python convert_mat_to_gen4.py 
python copy_paired_gen4.py 
python count_objects_gen4.pt 
python split_gen4_into_60s.py 
python split_gen4_60s_train_eval_test.py
python3 scripts/data/psee_to_frames.py data/gen4 data/gen1/gen1_preproc_npz -n 10 --mangling rvt-gen1 -d 50ms --workers 30
```
